```r
install.packages("tm")
install.packages("ColorBrewer")
library(tm)
library(wordcloud)
library(RColorBrewer)
```

```r
filesname = "Lab4_final_tweet_corpus.txt"
fileText=readLines(filesname)
```

```r
corp <- Corpus(VectorSource(fileText))
```

```r
map_data<-tm_map(corp,stripWhitespace)
map_data<-tm_map(map_data,tolower)
map_data<-tm_map(map_data,removeNumbers)
```

```r
map_data<-tm_map(map_data,removePunctuation)
map_data<-tm_map(map_data,removeWords, stopwords("english"))
```
